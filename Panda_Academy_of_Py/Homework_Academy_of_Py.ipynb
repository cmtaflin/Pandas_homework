```python
import pandas as pd
student_csv_pd = File =("Resources/students_complete.csv")
student_data =pd.read_csv(student_csv_pd)

school_csv_pd= File =("Resources/schools_complete.csv")
school_data= pd.read_csv(school_csv_pd)
```

```python
### Create Bins and create new columns with Pass/Fail
reading_bins=[0,69,100]
math_bins=[0,69,100]
overall_bins=[0,69,100]

#Labels
group_names=["Fail","Pass"]
Pass_name=[0,1]
```

```python
###### Pass / Fail Module  ########

#Reading Pass/Fail and Pass Count
student_data["Reading Pass/Fail"]=pd.cut(student_data["reading_score"],reading_bins,labels=group_names)
student_data["Reading Pass Count"]=pd.cut(student_data["reading_score"],reading_bins,labels=Pass_name)
student_data["Reading Pass Count"]=student_data["Reading Pass Count"].astype('float64')


##Math Pass/Fail and Pass Count
student_data["Math Pass/Fail"]=pd.cut(student_data["math_score"],math_bins,labels=group_names)
student_data["Math Pass Count"]=pd.cut(student_data["math_score"],math_bins,labels=Pass_name)
student_data["Math Pass Count"]=student_data["Math Pass Count"].astype('float64')

####################################
#student_data.columns
```

```python
###DISTRICT SUMMARY###
#Answer metrics with no Frame
total_schools=school_data["School ID"].count()
total_students=student_data["Student ID"].count()
total_budget=school_data["budget"].sum()
per_student_budget=total_schools/total_budget
school_data.head()
```

```python
### Merge School and Student data    ######
data_merged=pd.merge(student_data,school_data,how="left", on=["school_name","school_name"])

#data_merged.columns
```

```python
### District Counters
Student_counter=data_merged["size"]/data_merged["size"]
data_merged['Students']=Student_counter


Budget_per_student=data_merged["budget"]/data_merged["size"]
data_merged['Budget divided by student']=Budget_per_student

data_merged.head()
```

```python
#All Schools(tot)SCORE:Total Math,Reading with TOTAL avg Math and reading

#Total Math Scores
total_math_scores=student_data["math_score"].sum()
tot_avg_math_score=total_math_scores/total_students
student_data['Tot_Avg_Math']=tot_avg_math_score

#Total Reading Scores
total_read_scores=student_data["reading_score"].sum()
tot_avg_read_score=total_read_scores/total_students
student_data['Tot_Avg_Read']=tot_avg_read_score

#Percentage of students that passed math
total_math_pass=student_data.loc[student_data["Math Pass/Fail"]=="Pass","Math Pass/Fail"]
count_math_pass=total_math_pass.count()
tot_pct_math_pass=(count_math_pass/total_students)*100
student_data['Tot_Math_Pct_Pass']=tot_pct_math_pass

#Percentage of students that passed Reading
total_read_pass=student_data.loc[student_data["Reading Pass/Fail"]=="Pass","Reading Pass/Fail"]
count_read_pass=total_read_pass.count()
tot_pct_read_pass=(count_read_pass/total_students)*100
student_data['Tot_Pct_Pass']=tot_pct_read_pass

#Overall_pass
tot_pct_overeall_pass=(tot_pct_math_pass+ tot_pct_read_pass)/2
student_data['Tot_Overall_Pass_rate']=tot_pct_overeall_pass
#pct_overeall_pass
```

```python
######################################################         ANSWER 1             ####################################

#Set up summary Dataframe to print metrics from above
frame_df=pd.DataFrame({"Total Schools":[total_schools],"Total Students":[total_students],"Total Budget":[total_budget],
                      "Average Math Score":[tot_avg_math_score],"Average Reading Score":[tot_avg_read_score],
                      "% Passing Math":[tot_pct_math_pass],"% Passing Reading":[tot_pct_read_pass],
                       "Overall_Pass_rate":[tot_pct_overeall_pass]})

frame_df["Total Students"] = frame_df["Total Students"].astype(float).map("{:,.0f}".format)
frame_df["Total Budget"] = frame_df["Total Budget"].astype(float).map("${:,.2f}".format)
frame_df

#Format Columns
###Need to look into later
########################################################################################################################
```

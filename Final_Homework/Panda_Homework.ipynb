```python
import pandas as pd
student_csv_pd = ("Resources/students_complete.csv")
student_data =pd.read_csv(student_csv_pd)

school_csv_pd =("Resources/schools_complete.csv")
school_data= pd.read_csv(school_csv_pd)

# Combine the data into a single dataset
stu_sch_data= pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
###DISTRICT SUMMARY###
#Answer metrics with no Frame

#Total Schooos
total_schools=school_data["School ID"].count()

#Total Students
total_students=student_data["Student ID"].count()

#Total Budget
total_budget=school_data["budget"].sum()

#Cost per Student
per_student_budget=total_schools/total_budget
```

```python
### Create Bins and create new columns with Pass/Fail
reading_bins=[0,69,100]
math_bins=[0,69,100]
overall_bins=[0,69,100]

#Labels
group_names=["Fail","Pass"]
Pass_name=[0,1]

###### Pass / Fail Module  ########

#Reading Pass/Fail and Pass Count
stu_sch_data["Reading Pass/Fail"]=pd.cut(stu_sch_data["reading_score"],reading_bins,labels=group_names)
stu_sch_data["Reading Pass Count"]=pd.cut(stu_sch_data["reading_score"],reading_bins,labels=Pass_name)
stu_sch_data["Reading Pass Count"]=stu_sch_data["Reading Pass Count"].astype('float64')


##Math Pass/Fail and Pass Count
stu_sch_data["Math Pass/Fail"]=pd.cut(stu_sch_data["math_score"],math_bins,labels=group_names)
stu_sch_data["Math Pass Count"]=pd.cut(stu_sch_data["math_score"],math_bins,labels=Pass_name)
stu_sch_data["Math Pass Count"]=stu_sch_data["Math Pass Count"].astype('float64')
```

```python
#All Schools(tot)SCORE:Total Math,Reading with TOTAL avg Math and reading

#Total Math Scores
total_math_scores=stu_sch_data["math_score"].sum()
tot_avg_math_score=total_math_scores/total_students
stu_sch_data['Tot_Avg_Math']=tot_avg_math_score

#Total Reading Scores
total_read_scores=stu_sch_data["reading_score"].sum()
tot_avg_read_score=total_read_scores/total_students
stu_sch_data['Tot_Avg_Read']=tot_avg_read_score

#Percentage of students that passed math
total_math_pass=stu_sch_data.loc[stu_sch_data["Math Pass/Fail"]=="Pass","Math Pass/Fail"]
count_math_pass=total_math_pass.count()
tot_pct_math_pass=(count_math_pass/total_students)*100
stu_sch_data['Tot_Math_Pct_Pass']=tot_pct_math_pass

#Percentage of students that passed Reading
total_read_pass=stu_sch_data.loc[stu_sch_data["Reading Pass/Fail"]=="Pass","Reading Pass/Fail"]
count_read_pass=total_read_pass.count()
tot_pct_read_pass=(count_read_pass/total_students)*100
stu_sch_data['Tot_Pct_Pass']=tot_pct_read_pass

#Overall_pass
tot_pct_overeall_pass=(tot_pct_math_pass+ tot_pct_read_pass)/2
stu_sch_data['Tot_Overall_Pass_rate']=tot_pct_overeall_pass
```

```python
######################################################         ANSWER 1             ####################################

#Set up summary Dataframe to print metrics from above
Answer1_df=pd.DataFrame({"Total Schools":[total_schools],"Total Students":[total_students],"Total Budget":[total_budget],
                      "Average Math Score":[tot_avg_math_score],"Average Reading Score":[tot_avg_read_score],
                      "% Passing Math":[tot_pct_math_pass],"% Passing Reading":[tot_pct_read_pass],
                       "Overall_Pass_rate":[tot_pct_overeall_pass]})

Answer1_df["Total Students"] = Answer1_df["Total Students"].astype(float).map("{:,.0f}".format)
Answer1_df["Total Budget"] = Answer1_df["Total Budget"].astype(float).map("${:,.2f}".format)
Answer1_df
```
